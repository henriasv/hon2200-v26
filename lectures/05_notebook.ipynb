```python
import numpy as np
import matplotlib.pyplot as plt 
np.random.seed(0)
np.random.random()

Cs = [0, 20, -8, -10.7, 1.2, 1.6, -0.15, -0.05]

def f(x, Cs): 
    return sum([c*x**i for i, c in enumerate(Cs)])
N = 100
sigma = 7

x = np.linspace(-2.5, 2.5, N)
y = f(x, Cs) + np.random.normal(0, sigma, N)

plt.plot(x, y, 'o')
```

```python
stokkede_tall = np.random.permutation(100)
#stokkede_tall = np.arange(100)
idx = np.array_split(stokkede_tall, 5)

for i in range(len(idx)):
    plt.plot(x[idx[i]], y[idx[i]], 'o')
```
